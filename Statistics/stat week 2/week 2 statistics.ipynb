```python
import pandas as pd

# Suppose you have a DataFrame 'df' with columns 'Species', 'Sex', 'Island' and 'Culmen_Depth'
df = pd.DataFrame({
    'Species': ['Adelie', 'Gentoo', 'Chinstrap', 'Adelie', 'Gentoo'],
    'Sex': ['male', 'female', 'female', 'male', 'female'],
    'Island': ['Island1', 'Biscoe', 'Island3', 'Island1', 'Biscoe'],
    'Culmen_Depth': [18.7, 14.3, 18.1, 18.4, 14.2]
})

# Filter the DataFrame for Gentoo females on Biscoe island
df_gentoo_females_biscoe = df[(df['Species'] == 'Gentoo') & (df['Sex'] == 'female') & (df['Island'] == 'Biscoe')]

# Calculate the average culmen depth
average_culmen_depth_gentoo_females_biscoe = df_gentoo_females_biscoe['Culmen_Depth'].mean()

# Filter the DataFrame for all females
df_females = df[df['Sex'] == 'female']

# Group by species and calculate average culmen depth
average_culmen_depth_females = df_females.groupby('Species')['Culmen_Depth'].mean()

# Check if the average culmen depth of Gentoo females on Biscoe island is the least among all female species
is_least = average_culmen_depth_gentoo_females_biscoe == average_culmen_depth_females.min()

print(f"The average culmen depth of Gentoo females on Biscoe island is the least among all female species: {is_least}")

# Filter the DataFrame for Adelie males
df_adelie_males = df[(df['Species'] == 'Adelie') & (df['Sex'] == 'male')]

# Group by island and calculate average culmen depth
average_culmen_depth_adelie_males = df_adelie_males.groupby('Island')['Culmen_Depth'].mean()

# Check if the average culmen depths of Adelie males on the three islands are different
is_different = len(average_culmen_depth_adelie_males.unique()) == len(average_culmen_depth_adelie_males)

print(f"Average culmen depths of Adelie males on the three islands are different: {is_different}")
```

```python
import pandas as pd

# Suppose you have a DataFrame 'df' with columns 'Species', 'Sex' and 'Flipper_Length'
df = pd.DataFrame({
    'Species': ['Adelie', 'Gentoo', 'Chinstrap', 'Adelie', 'Gentoo'],
    'Sex': ['male', 'female', 'female', 'male', 'female'],
    'Flipper_Length': [181, 210, 191, 186, 215]
})

# Filter the DataFrame for females
df_females = df[df['Sex'] == 'female']

# Group by species and calculate average flipper length
average_flipper_length_females = df_females.groupby('Species')['Flipper_Length'].mean()

# Find the species with the highest average flipper length
highest_average_flipper_length_species = average_flipper_length_females.idxmax()

print(f"The species of penguin with the highest average flipper length among females is {highest_average_flipper_length_species}")
```
